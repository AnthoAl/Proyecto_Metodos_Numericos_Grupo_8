Como ya podemos visualizar de manera aproximada cuál puede ser la función que sirve como frontera de decisión, procedemos a identificar trazar puntos que estén encima de la frontera de decisión para luego, conectar esos puntos y así obtener la función objetivo.

```python
# BLACKBOX S - CÓDIGO CORREGIDO PARA EL MUESTREO DE BISECCIÓN
import numpy as np

# --- PARÁMETROS DE MUESTREO ---
x1_values = np.linspace(0.00, 1.5, 50)
x2_values = np.linspace(-1, 1.5, 20)
tolerancia = 0.00001
frontera_superior = []
frontera_inferior = []

print("Iniciando muestreo con bisección y estandarización de límites...")

for x1 in x1_values:
    
    for x2 in x2_values:
        y = predict_point(model, x1, x2)
        if y == 1:
            x2_low_sup = x2
            break
    
    # Frontera superior - Cambio de 1 a 0
    x2_high_sup = 1.5

    if predict_point(model, x1, x2_high_sup) == 0.0:
        low_sup, high_sup = x2_low_sup, x2_high_sup
        while abs(high_sup - low_sup) > tolerancia:
            x2_mid = (low_sup + high_sup) / 2
            if predict_point(model, x1, x2_mid) == 1.0:
                low_sup = x2_mid
            else:
                high_sup = x2_mid
        frontera_superior.append((x1, (low_sup + high_sup) / 2))
    
    # Frontera superior - Cambio de 0 a 1
    x2_high_inf = x2 
    x2_low_inf = -1.5
    
    if predict_point(model, x1, x2_low_inf) == 0.0:
        # Aplicar Bisección Inferior
        low_inf, high_inf = x2_low_inf, x2_high_inf
        while abs(high_inf - low_inf) > tolerancia:
            x2_mid = (low_inf + high_inf) / 2
            if predict_point(model, x1, x2_mid) == 0.0:
                low_inf = x2_mid
            else:
                high_inf = x2_mid
        frontera_inferior.append((x1, (low_inf + high_inf) / 2))

print(f"Puntos Frontera Superior: {len(frontera_superior)}")
print(f"Puntos Frontera Inferior: {len(frontera_inferior)}")
```

```python
# Método de Ajuste de Curvas - Frontera Inferior (Sinc Amortiguado) Algoritmo de levenberg-marquardt

from scipy.optimize import curve_fit

x1_inf_data = np.array([p[0] for p in frontera_inferior])
x2_inf_data = np.array([p[1] for p in frontera_inferior])

def sinc_model(x, A, B, C, D, E):
    """Modelo de curva: A / (x + B) * sin(C*x + D) + E"""
    # Evitar división por cero o valores cercanos
    denominator = x + B
    denominator[denominator < 0.01] = 0.01 
    
    # El amortiguamiento puede ser muy agresivo con x1+B, usaremos una forma más suave
    # Basado en la forma de tu gráfica, la amortiguación es RÁPIDA y luego se vuelve constante.
    
    # Si la frecuencia real es 10x, probemos con:
    return A * np.sin(C * x + D) / (C * x + 0.1) + E
    
# Usaremos una versión simplificada, ya que la anterior diverge:
def custom_damped_model(x, A, B, C, D):
    """Modelo adaptado: A * sin(B*x + C) / (x + 0.1) + D """
    return A * np.sin(B * x + C) / (x + 0.1) + D

# --- RE-AJUSTE CON MODELO SINC AMORTIGUADO ---
p0_inf_final = [0.7, 8.0, 0.0, 0.05] 

try:
    popt_inf, pcov_inf = curve_fit(
        custom_damped_model, 
        x1_inf_data, 
        x2_inf_data, 
        p0=p0_inf_final, 
        maxfev=5000 # Aumentar iteraciones para convergencia no lineal
    )
    A_inf, B_inf, C_inf, D_inf = popt_inf

    # Validar el nuevo punto inicial (x1=0)
    x2_pred_at_zero = custom_damped_model(0.01, *popt_inf) # Usamos 0.01 para evitar división por cero

    print("\n--- RESULTADOS DEL RE-AJUSTE FINAL (FRONTERA INFERIOR) ---")
    print(f"Nueva Relación Analítica (Sinc): x2 = {A_inf:.4f} * sin({B_inf:.4f} * x1 + {C_inf:.4f}) / (x1 + 0.1) + {D_inf:.4f}")
    print(f"Valor predicho en x1=0.01: x2 ≈ {x2_pred_at_zero:.4f}")

    # Calcular MSE (Prueba de Corrección)
    x2_pred = custom_damped_model(x1_inf_data, *popt_inf)
    mse = np.mean((x2_inf - x2_pred)**2)
    print(f"Error Cuadrático Medio (MSE): {mse:.10f}")
    
except RuntimeError as e:
    print(f"\nError en el ajuste final. La función sinc amortiguada es numéricamente inestable.")
    print("Se requiere una segmentación del modelo: Coseno puro [0, 0.9] y Constante [0.9, 1.5].")
```

```python
# Método 2 - Algoritmo de GAUSS NEWTON

import numpy as np

# Modelo Analítico (Sinc amortiguado)
def custom_damped_model(x, A, B, C, D):
    # Usamos x + 0.1 para evitar la singularidad en x=0, como en tu análisis.
    return A * np.sin(B * x + C) / (x + 0.1) + D

# Función de Residuos (Error)
def residual(beta, x_data, y_data):
    A, B, C, D = beta
    return y_data - custom_damped_model(x_data, A, B, C, D)

def jacobian(beta, x_data):
    A, B, C, D = beta
    
    # Calcular las partes comunes
    den = x_data + 0.1
    sin_term = np.sin(B * x_data + C)
    cos_term = np.cos(B * x_data + C)
    
    # Derivadas parciales (columnas de la matriz J)
    
    # d/dA
    J_A = sin_term / den
    
    # d/dB
    J_B = A * x_data * cos_term / den
    
    # d/dC
    J_C = A * cos_term / den
    
    # d/dD
    J_D = np.ones_like(x_data)
    
    # Apilar las columnas para formar la matriz Jacobiana
    return np.vstack((J_A, J_B, J_C, J_D)).T

def gauss_newton(x_data, y_data, beta_initial, max_iter=100, tolerance=1e-6):
    
    beta = np.array(beta_initial)
    
    for i in range(max_iter):
        
        # 1. Calcular el vector de residuos (r)
        r = residual(beta, x_data, y_data)
        
        # 2. Calcular la Matriz Jacobiana (J)
        J = jacobian(beta, x_data)
        
        # 3. Formar el lado izquierdo (J^T J) y el lado derecho (J^T r) del sistema
        JTJ = J.T @ J
        JTr = J.T @ r
        
        # 4. Resolver el sistema para encontrar el paso (Delta_beta)
        # Usamos np.linalg.lstsq para mayor estabilidad numérica que np.linalg.solve
        try:
            delta_beta = np.linalg.lstsq(JTJ, JTr, rcond=None)[0]
        except np.linalg.LinAlgError:
            print("Advertencia: Matriz JTJ singular. Deteniendo el algoritmo.")
            break
            
        # 5. Actualizar los parámetros
        beta_new = beta + delta_beta
        
        # 6. Criterio de Convergencia
        error = np.linalg.norm(delta_beta)
        if error < tolerance:
            print(f"Convergencia lograda en {i+1} iteraciones.")
            break
            
        beta = beta_new
        
        if i == max_iter - 1:
            print("Advertencia: Máximo de iteraciones alcanzado.")

    return beta

# --- SIMULACIÓN DE DATOS (Reemplaza esto con tus datos reales) ---
# Simular parámetros ajustados por Levenberg-Marquardt
# A=0.7, B=8.0, C=0.0, D=0.05
# x1_inf_data = np.linspace(0.01, 1.5, 100)
# x2_inf_data = 0.7 * np.sin(8.0 * x1_inf_data) / (x1_inf_data + 0.1) + 0.05 + np.random.normal(0, 0.01, 100)
# -----------------------------------------------------------------

# Estimación Inicial (p0):
beta_initial = [0.6, 7.5, 0.1, 0.1] 

# Parámetros óptimos encontrados por Gauss-Newton
beta_opt_gn = gauss_newton(x1_inf_data, x2_inf_data, beta_initial)

print("\nParámetros obtenidos por Gauss-Newton (A, B, C, D):")
print(beta_opt_gn)

A_infg, B_infg, C_infg, D_infg = beta_opt_gn

x2_pred_at_zero = custom_damped_model(0.01, *beta_opt_gn) # Usamos 0.01 para evitar división por cero

print("\n--- RESULTADOS DEL RE-AJUSTE FINAL (FRONTERA INFERIOR) con GAUSS NEWTON---")
print(f"Nueva Relación Analítica (Sinc): x2 = {A_infg:.4f} * sin({B_infg:.4f} * x1 + {C_infg:.4f}) / (x1 + 0.1) + {D_infg:.4f}")
print(f"Valor predicho en x1=0.01: x2 ≈ {x2_pred_at_zero:.4f}")

# Calcular MSE (Prueba de Corrección)
x2_pred = custom_damped_model(x1_inf_data, *popt_inf)
mse = np.mean((x2_inf - x2_pred)**2)
print(f"Error Cuadrático Medio (MSE): {mse:.10f}")
# Validar el nuevo punto inicial (x1=0)
   

# Comparación: Debes comparar estos parámetros con los obtenidos por curve_fit 
# y calcular el MSE para demostrar que son numéricamente equivalentes.
```
