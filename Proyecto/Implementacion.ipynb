Código para realizar el muestreo de los puntos ubicados en las fronteras de decisión del modelo a través del método de bisección

```python
import numpy as np

x1_values = np.linspace(0.00, 1.5, 50)
x2_values = np.linspace(-1, 1.5, 20)
tolerancia = 0.00001
frontera_superior = []
frontera_inferior = []

for x1 in x1_values:
    
    for x2 in x2_values:
        y = predict_point(model, x1, x2)
        if y == 1:
            x2_lim_bajo_sup = x2
            break
    
    # Frontera superior - Cambio de 1 a 0
    x2_lim_alto_sup = 1.5

    if predict_point(model, x1, x2_lim_alto_sup) == 0.0:
        lim_bajo_sup, lim_alto_sup = x2_lim_bajo_sup, x2_lim_alto_sup
        while abs(lim_alto_sup - lim_bajo_sup) > tolerancia:
            x2_medio = (lim_bajo_sup + lim_alto_sup) / 2
            if predict_point(model, x1, x2_medio) == 1.0:
                lim_bajo_sup = x2_medio
            else:
                lim_alto_sup = x2_medio
        frontera_superior.append((x1, (lim_bajo_sup + lim_alto_sup) / 2))
    
    # Frontera inferior - Cambio de 0 a 1
    x2_lim_alto_inf = x2 
    x2_lim_bajo_inf = -1.5
    
    if predict_point(model, x1, x2_lim_bajo_inf) == 0.0:
        lim_bajo_inf, lim_alto_inf = x2_lim_bajo_inf, x2_lim_alto_inf
        while abs(lim_alto_inf - lim_bajo_inf) > tolerancia:
            x2_medio = (lim_bajo_inf + lim_alto_inf) / 2
            if predict_point(model, x1, x2_medio) == 0.0:
                lim_bajo_inf = x2_medio
            else:
                lim_alto_inf = x2_medio
        frontera_inferior.append((x1, (lim_bajo_inf + lim_alto_inf) / 2))

print(f"Puntos Frontera Superior: {len(frontera_superior)}")
print(f"Puntos Frontera Inferior: {len(frontera_inferior)}")
```

Algoritmo de ajuste de curvas para la frontera inferior utilizando el método de Levenberg-Marquardt

```python
from scipy.optimize import curve_fit
import time

x1_inf_data = np.array([p[0] for p in frontera_inferior])
x2_inf_data = np.array([p[1] for p in frontera_inferior])
    
def modelo_regresion_nolineal(x, A, B, C, D):
    return A * np.sin(B * x + C) / (x + 0.1) + D

p0_inf_final = [0.7, 8.0, 0.0, 0.05] 

start_time = time.perf_counter()


popt_inf, pcov_inf = curve_fit(
    modelo_regresion_nolineal, 
    x1_inf_data, 
    x2_inf_data, 
    p0=p0_inf_final, 
    maxfev=5000
)

end_time = time.perf_counter()

# Cálculo del tiempo total de ejecución
elapsed_time = end_time - start_time
A_inf, B_inf, C_inf, D_inf = popt_inf
x2_pred_at_zero = modelo_regresion_nolineal(0.01, *popt_inf) 

print(f"Tiempo Total de Ejecución: {elapsed_time:.6f} segundos")

print("\nParámetros obtenidos por Levenberg-Maequardt (A, B, C, D):")
print(popt_inf)

print("\n--- RESULTADOS ---")
print(f"Relación Analítica: x2 = {A_inf:.4f} * sin({B_inf:.4f} * x1 + {C_inf:.4f}) / (x1 + 0.1) + {D_inf:.4f}")
print(f"Ejemplo del valor predicho para x1=0.01: x2 ≈ {x2_pred_at_zero:.4f}")
```

Algoritmo de ajuste de curvas para la frontera inferior utilizando el método de Gauss-Newton

```python
import numpy as np
import time

def modelo_regresion_nolineal(x, A, B, C, D):
    return A * np.sin(B * x + C) / (x + 0.1) + D

# Función de Residuos (Error)
def residual(beta, x_data, y_data):
    A, B, C, D = beta
    return y_data - modelo_regresion_nolineal(x_data, A, B, C, D)

def jacobian(beta, x_data):
    A, B, C, D = beta
    
    den = x_data + 0.1
    sin_term = np.sin(B * x_data + C)
    cos_term = np.cos(B * x_data + C)
    
    J_A = sin_term / den
    
    J_B = A * x_data * cos_term / den
    
    J_C = A * cos_term / den
    
    J_D = np.ones_like(x_data)
    
    return np.vstack((J_A, J_B, J_C, J_D)).T

def gauss_newton(x_data, y_data, beta_initial, max_iter=100, tolerance=1e-6):
    
    beta = np.array(beta_initial)
    
    start_time = time.perf_counter()

    for i in range(max_iter):
        
        r = residual(beta, x_data, y_data)
        
        J = jacobian(beta, x_data)
        
        JTJ = J.T @ J
        JTr = J.T @ r
        
        try:
            delta_beta = np.linalg.lstsq(JTJ, JTr, rcond=None)[0]
        except np.linalg.LinAlgError:
            print("Advertencia: Matriz JTJ singular. Deteniendo el algoritmo.")
            break
            
        beta_new = beta + delta_beta
        
        error = np.linalg.norm(delta_beta)
        if error < tolerance:
            print(f"Convergencia lograda en {i+1} iteraciones.")
            break
            
        beta = beta_new
        
        if i == max_iter - 1:
            print("Advertencia: Máximo de iteraciones alcanzado.")
    
    end_time = time.perf_counter()
    elapsed_time = end_time - start_time
    
    return beta, elapsed_time


beta_initial = [0.6, 7.5, 0.1, 0.1] 

# Parámetros óptimos encontrados por Gauss-Newton
beta_opt_gn, time_gn = gauss_newton(x1_inf_data, x2_inf_data, beta_initial)

A_infg, B_infg, C_infg, D_infg = beta_opt_gn

print("\nParámetros obtenidos por Gauss-Newton (A, B, C, D):")
print(beta_opt_gn)
print(f"Tiempo Total de Ejecución: {time_gn:.6f} segundos")

x2_pred_at_zero = modelo_regresion_nolineal(0.01, *beta_opt_gn)

print("\n--- RESULTADOS ---")
print(f"Relación Analítica: x2 = {A_infg:.4f} * sin({B_infg:.4f} * x1 + {C_infg:.4f}) / (x1 + 0.1) + {D_infg:.4f}")
print(f"Ejemplo del valor predicho en x1=0.01: x2 ≈ {x2_pred_at_zero:.4f}")
```

Generación de la gráfica comparativa entre el modelo ajustado y la función original

```python
import numpy as np
import matplotlib.pyplot as plt

def fun_original(x):
    return np.where(
        np.isclose(x, 0.0),
        1.0,
        np.sin(10 * x) / (10 * x)
    )

def fun_levenvergh(x):
    return A_inf * np.sin(B_inf * x + C_inf) / (x + 0.1) + D_inf

def fun_gauss(x):
    return A_infg * np.sin(B_infg * x + C_infg) / (x + 0.1) + D_infg

x_values = np.linspace(-5, 5, 400)

y_f = [fun_levenvergh(x) for x in x_values]
y_f1 = [fun_original(x) for x in x_values]
y_f2 = [fun_gauss(x) for x in x_values]

plt.figure(figsize=(10, 6))

plt.plot(x_values, y_f, label=f"x2 = ({A_inf:.4f} * sin({B_inf:.4f} * x1 + {C_inf:.4f})) / (x1 + 0.1) + {D_inf:.4f}", color='blue')
plt.plot(x_values, y_f1, label='$\\frac{\\sin{10x}}{10x}$', color='red')
plt.plot(x_values, y_f2, label=f"x2 = ({A_infg:.4f} * sin({B_infg:.4f} * x1 + {C_infg:.4f})) / (x1 + 0.1) + {D_infg:.4f}", color='orange', linestyle = 'dashed')

plt.title('Comparación de Modelos Ajustados vs Función Original')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()

plt.xlim(0,5.1)
plt.ylim(-1,1.5)

plt.show()
```

```python
#Calcular errores para el método de Levenberg-Marquardt

mselev = 0.0
msegauss = 0.0

for x1 in x1_inf_data:

    x2_real = fun_original(x1_inf_data)
    x2_pred_lev = fun_levenvergh(x1_inf_data)
    mselev += np.mean((x2_pred_lev - x2_real)**2)

    x2_pred_gauss = fun_gauss(x1_inf_data)
    msegauss += np.mean((x2_pred_gauss - x2_real)**2)

mselev /= len(x1_inf_data)
msegauss /= len(x1_inf_data)

print(f"Error Cuadrático Medio para Levenberg (MSE): {mselev:.10f}")

print(f"Error Cuadrático Medio para Gauss (MSE): {msegauss:.10f}")
```
